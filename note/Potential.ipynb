```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-0.1, 1.1, 0.01)
```

```python
def gen_f(a):
    def f(x):
        return 1 / (1-2*a) * (3*x**4 - 4*(a + 1)*x**3 + 6*a*x**2)
    return f
```

```python
a_vec = np.random.uniform(0,1/2, 10)
```

```python
fs = [gen_f(a) for a in a_vec]
ys = [f(x) for f in fs]
```

```python
plt.figure(figsize=(10,6), dpi=150)
for y in ys:
    plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation
def ode_system(t, y):
    x, v = y  # y contains x and x'
    a = 0.4
    x_double_prime = -0.4 * v - 30 * x * (x - 0.4) * (x - 1)
    return [v, x_double_prime]

# Initial conditions
x0 = -0.5  # Example initial condition
v0 = 0    # Given initial condition for x'
initial_conditions = [x0, v0]

# Time span (from 0 to some large time, e.g., 50)
t_span = [0, 50]
t_eval = np.linspace(t_span[0], t_span[1], 1000)  # Time points at which to solve

# Solve the ODE
sol = solve_ivp(ode_system, t_span, initial_conditions, t_eval=t_eval)

# Plotting the solution
plt.plot(sol.t, sol.y[0])
plt.xlabel('Time')
plt.ylabel('x(t)')
plt.title('Numerical solution of the ODE')
plt.grid(True)
plt.show()
```
